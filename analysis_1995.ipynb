```python
import numpy as np
import matplotlib.pyplot as plt
from math import log
```

```python
N = np.load("analysis_1995/rate_per_19x5.npy").T
```

```python
plt.scatter(N[0], N[1])
```

```python
plt.scatter(N[0], N[2])
```

```python
print(np.corrcoef(N[0], N[1])[0,1])
print(np.corrcoef(N[0], N[2])[0,1])
```

```python
l = 30
r = 100
print(np.corrcoef(N[0, l:-r], N[1, l:-r])[0,1])
print(np.corrcoef(N[0, l:-r], N[2, l:-r])[0,1])
```

```python
x = N[0]
y = N[1]
z = np.polyfit(x, y, 2)

p = np.poly1d(z)
xp = np.linspace(-500, 4000, 100)
plt.vlines([300,3000], 0.2, 0.8)

plt.plot(x, y, '.', xp, p(xp), '-')
plt.show()
print(p)
print(np.corrcoef(x, y)[0,1])
```

```python
x = N[0]
y = N[2]
z = np.polyfit(x, y, 2)

p = np.poly1d(z)
xp = np.linspace(-500, 4000, 100)

plt.plot(x, y, '.', xp, p(xp), '-')
plt.show()
print(p)
print(np.corrcoef(x, y)[0,1])
```
